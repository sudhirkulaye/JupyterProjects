```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
from datetime import datetime

# Define the file paths for TCS, ticker2, ticker3, and ticker4
ticker1 = 'TCS'
ticker2 = 'WIPRO'  # Leave blank if not provided
ticker3 = ''  # Leave blank if not provided
ticker4 = ''  # Leave blank if not provided
ticker5 = ''  # Leave blank if not provided
ticker6 = ''  # Leave blank if not provided
# Start column and end column
startColumn = 1  # Default 1
endColumn = 14  # Default 14

DollarRevenue = 1
Employees = 79
OneMillionClients = 83
TenMillionClients = 85
FiftyMillionClients = 87
HundredMillionClients = 88
TotalCLients = 89
TTMDeals = 99


# Define the row number for the data to plot Technology Tickers
rowNumber1 = DollarRevenue  # Assuming you want to compare sales data
rowNumber2 = Employees  # Y-axis data for the second chart
rowNumber3 = OneMillionClients  # Y-axis data for the third chart
rowNumber4 = TenMillionClients  # Y-axis data for the fourth chart
rowNumber5 = FiftyMillionClients  # Y-axis data for the fifth chart
rowNumber6 = HundredMillionClients  # Y-axis data for the fifth chart
rowNumber7 = TotalCLients  # Y-axis data for the fifth chart
rowNumber8 = TTMDeals  # Y-axis data for the fifth chart

# Specify whether to plot a bar chart (True) or a line chart (False)
plot_bar_chart = False

file_path_ticker1 = r'C:\MyDocuments\03Business\05ResearchAndAnalysis\StockInvestments\QuarterResultsScreenerExcels\2025Q4\{}_FY25Q4.xlsx'.format(ticker1)

# Read the Excel file into a dataframe for ticker1
df_ticker1 = pd.read_excel(file_path_ticker1, sheet_name='SegmentAnalysis')

# Extract the label for the specified row (assuming all files have the same structure)
row_label1 = df_ticker1.iloc[rowNumber1, 0]
row_label2 = df_ticker1.iloc[rowNumber2, 0]
row_label3 = df_ticker1.iloc[rowNumber3, 0]
row_label4 = df_ticker1.iloc[rowNumber4, 0]
row_label5 = df_ticker1.iloc[rowNumber5, 0]
row_label6 = df_ticker1.iloc[rowNumber6, 0]
row_label7 = df_ticker1.iloc[rowNumber7, 0]
row_label8 = df_ticker1.iloc[rowNumber8, 0]

# Extract cell values for x-axis (dates) from the dataframe for ticker1
dates = df_ticker1.iloc[0, startColumn:endColumn-1]
dates = [str(date.date()) for date in dates]  # Convert dates to strings in desired format

# Get the latest quarter and convert it to the desired format
latest_quarter = pd.read_excel(file_path_ticker1, sheet_name='QuarterP&L').iloc[1, 17]
latest_quarter = str(latest_quarter.date())

# Append today's date to the dates list
dates.append(latest_quarter)

# Display the Markdown-formatted text
display(Markdown("## **Annual Segment Numbers**"))

# Plot the charts
for row_number, row_label in [(rowNumber1, row_label1), (rowNumber2, row_label2), (rowNumber3, row_label3), (rowNumber4, row_label4), (rowNumber5, row_label5), (rowNumber6, row_label6), (rowNumber7, row_label7), (rowNumber8, row_label8)]:
    plt.figure(figsize=(12, 8))  # Adjust the figure size here
    
    for i, ticker in enumerate([ticker1, ticker2, ticker3, ticker4, ticker5, ticker6]):
        if ticker == '' or ticker is None:
            continue
        
        file_path_ticker = r'C:\MyDocuments\03Business\05ResearchAndAnalysis\StockInvestments\QuarterResultsScreenerExcels\2025Q4\{}_FY25Q4.xlsx'.format(ticker)
        df_ticker = pd.read_excel(file_path_ticker, sheet_name='SegmentAnalysis')
        
        # Extract cell values for y-axis (sales numbers) from the dataframe for the current ticker
        #sales_numbers = df_ticker.iloc[row_number, startColumn:endColumn].astype(float)
        sales_numbers = df_ticker.iloc[row_number, startColumn:endColumn].fillna(0).astype(float)
        
        # Extracting only the last value from the series
        last_value = sales_numbers.iloc[-1]
        
        # Check the conditions and format last_value accordingly
        if last_value < 1:
            last_value_formatted = "{:.4f}".format(last_value)  # Format to 4 decimal places
        elif last_value > 100:
            last_value_formatted = "{:.0f}".format(last_value)  # Format to 0 decimal places
        else:
            last_value_formatted = "{:.2f}".format(last_value)  # Format to 2 decimal places
        
        # Plot the chart based on the selected chart type
        if plot_bar_chart:
            plt.bar(dates, sales_numbers, width=0.2, label=ticker)
        else:
            plt.plot(dates, sales_numbers, marker='o', linestyle='-', label=ticker)
        
        # Displaying only the last value on the chart with appropriate rounding
        if last_value >= 100:
            plt.text(dates[-1], last_value, str(round(last_value)), ha='right', va='bottom')
        else:
            plt.text(dates[-1], last_value, str(round(last_value, 3)), ha='right', va='bottom')
    
    plt.title('Annual ' + row_label + ' Comparison')
    plt.xlabel('FIN Year')
    plt.ylabel(row_label)
    plt.xticks(rotation=45, ha='right')  # Adjust rotation and alignment of x-axis labels
    plt.legend()  # Show legend
    plt.tight_layout()
    plt.show()
    
# Define the row numbers for each combination of Geographical and Segment Revenue for each ticker
combinations = {
    "North America Revenue": {
        ticker1: 43,
        ticker2: 43,
        ticker3: 46,
        ticker4: 43,
        ticker5: 43,
        ticker6: 43,
    },
    "North America Contribution%": {
        ticker1: 52,
        ticker2: 52,
        ticker3: 55,
        ticker4: 52,
        ticker5: 52,
        ticker6: 52,
    },
    "Europe Revenue": {
        ticker1: 46,
        ticker2: 44,
        ticker3: 47,
        ticker4: 46,
        ticker5: 44,
        ticker6: 44,
    },
    "Europe Contribution%": {
        ticker1: 55,
        ticker2: 53,
        ticker3: 56,
        ticker4: 55,
        ticker5: 53,
        ticker6: 53,
    },
    "BFSI Revenue": {
        ticker1: 2,
        ticker2: 2,
        ticker3: 2,
        ticker4: 2,
        ticker5: 2,
        ticker6: 5,
    },
    "BFSI Contribution%": {
        ticker1: 14,
        ticker2: 14,
        ticker3: 14,
        ticker4: 14,
        ticker5: 14,
        ticker6: 17,
    },
    "Retail Revenue": {
        ticker1: 3,
        ticker2: 3,
        ticker3: 5,
        ticker4: 3,
        ticker5: 5,
        ticker6: 6,
    },
    "Retail Contribution%": {
        ticker1: 15,
        ticker2: 15,
        ticker3: 17,
        ticker4: 15,
        ticker5: 17,
        ticker6: 18,
    },
    # Add more combinations as needed
}

# Display the Markdown-formatted text
display(Markdown("## **Annual Segment Numbers**"))

# Specify whether to plot a bar chart (True) or a line chart (False)
plot_bar_chart = False

# Loop through each combination
for combination_name, combination_rows in combinations.items():
    # Plot the charts for the current combination
    plt.figure(figsize=(12, 8))  # Adjust the figure size
    
    for ticker, row_number in combination_rows.items():
        if ticker == '' or ticker is None:
            continue
        
        # Get the file path for the current ticker
        file_path_ticker = r'C:\MyDocuments\03Business\05ResearchAndAnalysis\StockInvestments\QuarterResultsScreenerExcels\2025Q4\{}_FY25Q4.xlsx'.format(ticker)
        df_ticker = pd.read_excel(file_path_ticker, sheet_name='SegmentAnalysis')
        
        # Extract cell values for y-axis (Revenue)
        revenue = df_ticker.iloc[row_number, startColumn:endColumn].fillna(0).astype(float)
        row_name = df_ticker.iloc[row_number, 0]
        
        plt.plot(dates, revenue, marker='o', linestyle='-', label=ticker + '-' + row_name)
        
        # Displaying only the last value on the chart with appropriate rounding
        last_value = revenue.iloc[-1]
        if last_value < 1:
            last_value_formatted = "{:.4f}".format(last_value)  # Format to 4 decimal places
        elif last_value > 100:
            last_value_formatted = "{:.0f}".format(last_value)  # Format to 0 decimal places
        else:
            last_value_formatted = "{:.2f}".format(last_value)  # Format to 2 decimal places
        
        plt.text(dates[-1], last_value, last_value_formatted, ha='right', va='bottom')
    
    plt.title('Annual ' + combination_name + ' Comparison')
    plt.xlabel('FIN Year')
    plt.ylabel(combination_name)
    plt.xticks(rotation=45, ha='right')  # Adjust rotation and alignment of x-axis labels
    plt.legend()  # Show legend
    plt.tight_layout()
    plt.show()
```
