```python
import os
import pandas as pd

# Define the folder path
folder_path = r"C:\MyDocuments\03Business\05ResearchAndAnalysis\StockInvestments\QuarterResultsScreenerExcels\2026Q2"

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx") and not filename.startswith("~$"):
        file_path = os.path.join(folder_path, filename)
        try:
            # Read the Excel file with the specified sheet name
            df = pd.read_excel(file_path, sheet_name="Data Sheet", header=None)  # Set header=None
            # Extract the value from cell B1 (1st row, 1st column)
            value_B1 = df.iloc[0, 1]
            # Extract the ticker from the filename (excluding the ".xlsx" extension)
            ticker = filename.split("_")[0]
            # Check if the modified ticker matches the value of cell B1
            if ticker != value_B1:
                print(f"Filename: {filename}, Cell B1 Value: {value_B1}, Ticker: {ticker}")
        except Exception as e:
            print(f"Error reading {filename}: {e}")
```
