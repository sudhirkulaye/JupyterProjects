```python
# Initialize the OpenAI API client
openai.api_key = api_key

def generate_summary(text):
    # Call the OpenAI GPT-3.5 API to generate a summary
    response = openai.Completion.create(
        engine="davinci",  # Specify the engine for backward compatibility
        prompt=text,
        max_tokens=150,  # Adjust max_tokens as needed for longer text
        temperature=0.5,  # Adjust temperature for creativity vs. coherence balance
        top_p=1.0,  # Adjust top_p for diversity vs. quality balance
        frequency_penalty=0.0,
        presence_penalty=0.0,
    )
    # Extract and return the generated summary
    summary = response['choices'][0]['text'].strip()
    return summary

# Example text from PDF pages 1 to 4
pdf_text = """
Lorem ipsum dolor sit amet, consectetur adipiscing elit. Sed faucibus enim vel aliquam fringilla. Mauris auctor leo a risus vehicula, sit amet condimentum lectus commodo. Integer ultrices, purus vel sollicitudin hendrerit, libero sapien lobortis dui, at consequat felis tortor id lacus. Sed consequat felis at lacinia tristique. Donec non tempor mi. Integer id ullamcorper eros. Cras id velit nec ipsum dictum rutrum. Sed porttitor libero at lectus euismod dictum. Nam in commodo neque. Pellentesque habitant morbi tristique senectus et netus et malesuada fames ac turpis egestas. Phasellus euismod est magna, sit amet faucibus metus congue eu. Cras convallis efficitur velit, a laoreet nibh posuere at. Sed id vehicula leo, in laoreet orci. Morbi id mauris et ex laoreet auctor vel ac risus.

Nulla facilisi. Sed vehicula libero eget nibh dapibus, vitae blandit metus posuere. Nam suscipit dapibus orci, sit amet dapibus mi. Morbi bibendum felis ac suscipit commodo. Fusce at nunc vitae nisi ullamcorper ultrices eget eget est. Vivamus efficitur quam vitae libero consequat, nec consequat lorem elementum. Curabitur eget condimentum urna, non sodales tortor. Sed nec enim dui. Donec pretium, nulla a dictum lobortis, lacus purus rutrum lorem, sed pellentesque metus sem ut urna. Integer bibendum enim nec semper lacinia. Ut lacinia purus id posuere egestas. Maecenas rutrum tellus id magna ultricies, vel efficitur enim eleifend. Phasellus non ultricies ex. Donec sit amet eleifend justo. Ut vel ornare metus.

Etiam non odio sed nunc consequat convallis. Proin pharetra libero ut purus ultricies lacinia. Duis vitae nibh at ipsum pharetra vestibulum. Fusce rhoncus
"""

# Generate summary for the example text
summary = generate_summary(pdf_text)

# Print the summary
print("Summary:")
print(summary)
```
