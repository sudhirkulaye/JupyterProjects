```python
import os
import openpyxl
from datetime import datetime, timedelta

# Constants and variables
BASE_FOLDER = r"C:\MyDocuments\03Business\05ResearchAndAnalysis\StockInvestments\QuarterResultsScreenerExcels"
quarter_folder = "2025Q2"  # Change this variable when needed
file_name = ""  # Specify a file name (e.g., "example.xlsx") or leave blank to process all files
lastmodified = 0  # Number of days to filter by last modified date; 0 means all files

# Construct folder path
folder_path = os.path.join(BASE_FOLDER, quarter_folder)

# Output SQL file name
output_filename = datetime.now().strftime("%Y%m%d%H%M_DBInsertScript.sql")
output_file_path = os.path.join(folder_path, output_filename)

# Time filter for last modified files
time_filter = datetime.now() - timedelta(days=lastmodified)

# Initialize an empty list to collect SQL insert statements
sql_statements = []

# Process files
if file_name:  # Process only the specified file
    files_to_process = [file_name]
else:  # Process all files
    files_to_process = os.listdir(folder_path)

for file in files_to_process:
    file_path = os.path.join(folder_path, file)

    # Skip temporary files or ensure it matches modification criteria
    if file.startswith("~$") or not file.endswith((".xlsx", ".xls")):
        continue
    if not file_name and lastmodified > 0 and datetime.fromtimestamp(os.path.getmtime(file_path)) <= time_filter:
        continue

    print(f"Processing file: {file}")  # Log file being processed

    try:
        # Open the Excel file with openpyxl for precise control over cells
        workbook = openpyxl.load_workbook(file_path, data_only=True)
        if "DBInsert" not in workbook.sheetnames:
            print(f"DBInsert sheet not found in {file}")
            continue
        
        sheet = workbook["DBInsert"]
        
        # Iterate through rows starting from row 73 (index 72) in column A (index 0)
        for row in sheet.iter_rows(min_row=73, max_row=sheet.max_row, min_col=1, max_col=1):
            cell = row[0]
            value = cell.value
            # Skip blank, None, empty strings, or cells with "#N/A"
            if value and isinstance(value, str) and value.strip() and value.strip() != "#N/A":
                sql_statements.append(value.strip())
    
    except Exception as e:
        print(f"Error processing file {file}: {e}")

# Write collected SQL statements to the output file
if sql_statements:
    with open(output_file_path, "w") as sql_file:
        for statement in sql_statements:
            sql_file.write(f"{statement}\n")
    print(f"SQL file generated: {output_file_path}")
else:
    print("No SQL statements extracted.")
```

```python
import os
import win32com.client

# Folder containing the Excel files
BASE_FOLDER = r"C:\MyDocuments\03Business\05ResearchAndAnalysis\StockInvestments\QuarterResultsScreenerExcels"
quarter_folder = "2025Q2"  # Change this variable when needed
folder_path = os.path.join(BASE_FOLDER, quarter_folder)

# Initialize Excel application
excel = win32com.client.Dispatch("Excel.Application")
excel.Visible = False  # Keep Excel hidden during processing

# Process each Excel file
for file in os.listdir(folder_path):
    if file.startswith("~$") or not file.endswith((".xlsx", ".xls")):
        continue  # Skip temporary or non-Excel files

    file_path = os.path.join(folder_path, file)
    try:
        print(f"Processing file: {file}")
        workbook = excel.Workbooks.Open(file_path)
        workbook.Application.CalculateFull()  # Force recalculation of all formulas
        workbook.Save()  # Save the updated file
        workbook.Close()
    except Exception as e:
        print(f"Error processing file {file}: {e}")

# Quit Excel application
excel.Quit()
print("Recalculation and save completed for all files.")
```
