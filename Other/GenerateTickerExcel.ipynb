```python
import os
import pymysql
import pandas as pd
from openpyxl import Workbook

# Define the ticker as a constant
TICKER = 'BLUESTARCO'

# Database Connection Details
DB_CONFIG = {
    'host': 'localhost',
    'user': 'towdevuser',
    'password': 'Dev703',
    'database': 'timelineofwealth'
}

# Output directory
OUTPUT_DIR = r"C:\MyDocuments\03Business\05ResearchAndAnalysis\StockInvestments\QuarterResultsScreenerExcels\ChatGPT"

def generate_excel_from_db(ticker, output_file):
    # Check if file exists and delete it
    if os.path.exists(output_file):
        os.remove(output_file)
        print(f"Existing file '{output_file}' has been deleted.")

    # Establish database connection
    conn = pymysql.connect(**DB_CONFIG)
    
    # Define queries
    queries = {
        "QuarterlyP&L": f"SELECT * FROM stock_quarter WHERE ticker = '{ticker}'",
        "AnnualP&L": f"SELECT * FROM stock_pnl WHERE ticker = '{ticker}'",
        "Balancesheet": f"SELECT * FROM stock_balancesheet WHERE ticker = '{ticker}'",
        "Cashflow": f"SELECT * FROM stock_cashflow WHERE ticker = '{ticker}'",
        "AnalystReco": f"SELECT * FROM stock_analyst_reco WHERE ticker = '{ticker}'",
        "Valuation": f"SELECT * FROM stock_valuation WHERE ticker = '{ticker}'",
        "PriceHIstory": f"SELECT * FROM nse_price_history WHERE nse_ticker = '{ticker}'",
        "AnalystConCallHistory": None,  # No query; leave empty
        "AnnualReportsNotes": None      # No query; leave empty
    }

    # Create a new workbook
    wb = Workbook()
    # Remove the default sheet created by Workbook
    if "Sheet" in wb.sheetnames:
        wb.remove(wb["Sheet"])
    
    for sheet_name, query in queries.items():
        ws = wb.create_sheet(title=sheet_name)
        if query:
            try:
                # Execute the query and fetch data into a DataFrame
                df = pd.read_sql(query, conn)
                # Write the DataFrame to the worksheet
                for col_num, column in enumerate(df.columns, start=1):
                    ws.cell(row=1, column=col_num, value=column)  # Write column headers
                for row_num, row in enumerate(df.itertuples(index=False), start=2):
                    for col_num, value in enumerate(row, start=1):
                        ws.cell(row=row_num, column=col_num, value=value)
            except Exception as e:
                print(f"Error processing sheet {sheet_name}: {e}")
        else:
            # For empty sheets, just leave the headers blank
            print(f"No data for sheet {sheet_name}. Leaving it empty.")

    # Save the workbook
    wb.save(output_file)
    print(f"Excel file '{output_file}' generated successfully!")

    # Close database connection
    conn.close()

# Define the output file path
output_file = os.path.join(OUTPUT_DIR, f"{TICKER}_CHATGPT.xlsx")

# Ensure the output directory exists
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Generate the Excel file
generate_excel_from_db(TICKER, output_file)
```
