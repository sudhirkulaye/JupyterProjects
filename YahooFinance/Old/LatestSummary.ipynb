```python
import pandas as pd
import os
import re

# ================= CONFIG =================
FOLDER = r"C:\MyDocuments\03Business\05ResearchAndAnalysis\StockInvestments\ResearchReports\YahooConsensus"
MIN_ANALYSTS = 5      # configurable
TOP_N = 10            # configurable

# ================= FIND LATEST FILE =================
files = [
    f for f in os.listdir(FOLDER)
    if re.match(r"\d{8}_YahooConsensus\.csv", f)
]

latest_file = max(files)
path = os.path.join(FOLDER, latest_file)

print("Using file:", latest_file)

# ================= LOAD =================
df = pd.read_csv(path)

# ================= PREP =================
# Filter by analyst count
df = df[df["#Count"] >= MIN_ANALYSTS]

# ================= IDENTIFY Y1 & Y2 =================
fy_sorted = sorted(df["FIN Yr"].unique())
Y1 = fy_sorted[0]
Y2 = fy_sorted[1]

print("Y1:", Y1, "| Y2:", Y2)

# Keep one row per ticker (TP appreciation is same across FYs)
df_tp = (
    df.sort_values("FIN Yr")
      .drop_duplicates(subset=["YAHOO Ticker"], keep="first")
)

# ================= A) TOP MIN TP APPRECIATION =================
top_min_tp = (
    df_tp[["YAHOO Ticker", "TP_Appreciation_(Min)", "#Count"]]
    .sort_values("TP_Appreciation_(Min)", ascending=False)
    .head(TOP_N)
)

# ================= B) TOP AVG TP APPRECIATION =================
top_avg_tp = (
    df_tp[["YAHOO Ticker", "TP_Appreciation_(Avg)", "#Count"]]
    .sort_values("TP_Appreciation_(Avg)", ascending=False)
    .head(TOP_N)
)

# ================= C) TOP MAX TP APPRECIATION =================
top_max_tp = (
    df_tp[["YAHOO Ticker", "TP_Appreciation_(Max)", "#Count"]]
    .sort_values("TP_Appreciation_(Max)", ascending=False)
    .head(TOP_N)
)

# ================= D) Bottom MIN TP APPRECIATION =================
bottom_min_tp = (
    df_tp[["YAHOO Ticker", "TP_Appreciation_(Min)", "#Count"]]
    .sort_values("TP_Appreciation_(Min)", ascending=True)
    .head(TOP_N)
)

# ================= E) Bottom AVG TP APPRECIATION =================
bottom_avg_tp = (
    df_tp[["YAHOO Ticker", "TP_Appreciation_(Avg)", "#Count"]]
    .sort_values("TP_Appreciation_(Avg)", ascending=True)
    .head(TOP_N)
)

# ================= F) Bottom MAX TP APPRECIATION =================
bottom_max_tp = (
    df_tp[["YAHOO Ticker", "TP_Appreciation_(Max)", "#Count"]]
    .sort_values("TP_Appreciation_(Max)", ascending=True)
    .head(TOP_N)
)

# ================= DISPLAY =================
print("\nA) Top Min Target Price Appreciation")
print(top_min_tp.to_string(index=False))

print("\nB) Top Avg Target Price Appreciation")
print(top_avg_tp.to_string(index=False))

print("\nC) Top Max Target Price Appreciation")
print(top_max_tp.to_string(index=False))

# ================= DISPLAY =================
print("\nD) Bottom Min Target Price Appreciation")
print(bottom_min_tp.to_string(index=False))

print("\nE) Bottom Avg Target Price Appreciation")
print(bottom_avg_tp.to_string(index=False))

print("\nF) Bottom Max Target Price Appreciation")
print(bottom_max_tp.to_string(index=False))
```

```python
# Helper to select + rank
def top_rank(df, fy, metric, label):
    out = (
        df[df["FIN Yr"] == fy]
        .sort_values(metric, ascending=False)
        .head(TOP_N)
        [[
            "YAHOO Ticker",
            metric,
            "TP_Appreciation_(Min)",
            "TP_Appreciation_(Avg)",
            "TP_Appreciation_(Max)",
            "#Count"
        ]]
    )
    print(f"\n{label} ({fy})")
    print(out.to_string(index=False))

# ================= G) FY26(Y1) Min Rev Growth =================
top_rank(
    df,
    Y1,
    "Rev_Growth_(Min)",
    "G) Top Min Revenue Growth"
)

# ================= H) FY27(Y2) Min Rev Growth =================
top_rank(
    df,
    Y2,
    "Rev_Growth_(Min)",
    "H) Top Min Revenue Growth"
)

# ================= I) FY26(Y1) Avg Rev Growth =================
top_rank(
    df,
    Y1,
    "Rev_Growth_(Avg)",
    "I) Top Avg Revenue Growth"
)

# ================= J) FY27(Y2) Avg Rev Growth =================
top_rank(
    df,
    Y2,
    "Rev_Growth_(Avg)",
    "J) Top Avg Revenue Growth"
)

# ================= K) FY26(Y1) Max Rev Growth =================
top_rank(
    df,
    Y1,
    "Rev_Growth_(Max)",
    "K) Top Max Revenue Growth"
)
# ================= L) FY27(Y2) Max Rev Growth =================
top_rank(
    df,
    Y2,
    "Rev_Growth_(Max)",
    "L) Top Max Revenue Growth"
)
```

```python
# ================= M) FY26(Y1) Min EPS Growth =================
top_rank(
    df,
    Y1,
    "EPS_Growth_(Min)",
    "M) Top Min EPS Growth"
)

# ================= N) FY27(Y2) Min EPS Growth =================
top_rank(
    df,
    Y2,
    "EPS_Growth_(Min)",
    "N) Top Min EPS Growth"
)

# ================= O) FY26(Y1) Avg EPS Growth =================
top_rank(
    df,
    Y1,
    "EPS_Growth_(Avg)",
    "O) Top Avg EPS Growth"
)

# ================= P) FY27(Y2) Avg EPS Growth =================
top_rank(
    df,
    Y2,
    "EPS_Growth_(Avg)",
    "P) Top Avg EPS Growth"
)

# ================= Q) FY26(Y1) Max EPS Growth =================
top_rank(
    df,
    Y1,
    "EPS_Growth_(Max)",
    "Q) Top Max EPS Growth"
)

# ================= R) FY27(Y2) Max EPS Growth =================
top_rank(
    df,
    Y2,
    "EPS_Growth_(Max)",
    "R) Top Max EPS Growth"
)
```
