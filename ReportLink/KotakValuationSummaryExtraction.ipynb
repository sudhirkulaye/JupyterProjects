```python
import re
import csv
import pdfplumber
from datetime import datetime

# Base path for input and output files
base_path = "C:/MyDocuments/03Business/05ResearchAndAnalysis/StockInvestments/ResearchReports/CompanyResearchReports/KotakDaily"

# Input PDF file name (change this as needed)
input_pdf_file = "indiadaily18032025at_Valuation.pdf"  # Replace with your PDF file name

# Output CSV file name (dynamic based on today's date)
output_csv_file = datetime.now().strftime("%Y%m%d") + "_KotakValuationSummary.csv"

# Full paths for input and output files
pdf_path = f"{base_path}/{input_pdf_file}"
csv_path = f"{base_path}/{output_csv_file}"

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        # Extract only the last 5 pages (assuming valuation summary is at the end)
        for page in pdf.pages[-5:]:
            text += page.extract_text()
    return text

# Function to parse the text and extract relevant data
def parse_text_to_csv(text, csv_path):
    # Split the text into lines
    lines = text.splitlines()

    # Initialize a list to store rows
    rows = []

    # Iterate through the lines and extract relevant data
    for line in lines:
        # Use regex to match the data pattern
        match = re.match(
            r"([A-Za-z\s&\.-]+)\s+"  # Company name
            r"(SELL|ADD|BUY|REDUCE)\s+"  # Rating
            r"([\d,]+)\s+"  # CMP
            r"([\d,]+)\s+"  # Fair Value
            r"([\d-]+)\s+"  # Upside (%)
            r"([\d,]+)\s+"  # Market Cap (bn Rs.)
            r"([\d\.]+)\s+"  # Market Cap ($b)
            r"([\d,]+)\s+"  # O/S Shares
            r"([\d-]+)\s+"  # EPS_Y0
            r"([\d-]+)\s+"  # EPS_Y1
            r"([\d-]+)\s+"  # EPS_Y2
            r"([\d-]+)\s+"  # PE_Y0
            r"([\d-]+)\s+"  # PE_Y1
            r"([\d-]+)\s+"  # PE_Y2
            r"([\d\.]+)\s+"  # PB_Y0
            r"([\d\.]+)\s+"  # PB_Y1
            r"([\d\.]+)\s+"  # PB_Y2
            r"([\d-]+)\s+"  # EV/EBITDA_Y0
            r"([\d-]+)\s+"  # EV/EBITDA_Y1
            r"([\d-]+)\s+"  # EV/EBITDA_Y2
            r"([\d-]+)\s+"  # RoE_Y0
            r"([\d-]+)\s+"  # RoE_Y1
            r"([\d-]+)\s+"  # RoE_Y2
            r"([\d\.]+)\s+"  # DivYield_Y0
            r"([\d\.]+)\s+"  # DivYield_Y1
            r"([\d\.]+)\s+"  # DivYield_Y2
            r"([\d-]+)\s+"  # 3M-ADV-Traded ($mn)
            r"([\d-]+)",  # 3M-ADV-Delivered ($mn)
            line,
        )
        if match:
            rows.append(match.groups())

    # Write data to CSV
    with open(csv_path, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        # Write header
        writer.writerow([
            "Company", "Ratings", "CMP", "Fair Value", "Upside (%)", "MarketCap(bn Rs.)",
            "MarketCap($b)", "o/s shares", "EPS_Y0", "EPS_Y1", "EPS_Y2", "2yrsEPSg%",
            "PE_Y0", "PE_Y1", "PE_Y2", "PB_Y0", "PB_Y1", "PB_Y2", "EV/EBITDA_Y0",
            "EV/EBITDA_Y1", "EV/EBITDA_Y2", "RoE_Y0", "RoE_Y1", "RoE_Y2", "DivYield_Y0",
            "DivYield_Y1", "DivYield_Y2", "3M-ADV-Traded($mn)", "3M-ADV-Delivered($mn)"
        ])
        # Write each row
        for row in rows:
            # Compute 2yr EPS growth
            eps_y0 = float(row[8].replace(",", "")) if row[8] != "-" and row[8] else 0
            eps_y2 = float(row[10].replace(",", "")) if row[10] != "-" and row[10] else 0
            eps_growth = ((eps_y2 - eps_y0) / eps_y0) * 100 if eps_y0 != 0 else 0

            # Write the row to CSV
            writer.writerow([
                row[0].strip(), row[1], row[2], row[3], row[4], row[5], row[6], row[7],
                row[8], row[9], row[10], f"{eps_growth:.2f}", row[11], row[12], row[13],
                row[14], row[15], row[16], row[17], row[18], row[19], row[20], row[21],
                row[22], row[23], row[24], row[25], row[26], row[27]
            ])

# Main function
def main():
    # Extract text from PDF (only last 5 pages)
    text = extract_text_from_pdf(pdf_path)
    print("Extracted Text:\n", text)  # Debug: Print extracted text

    # Parse text and write to CSV
    parse_text_to_csv(text, csv_path)
    print(f"Data extracted and saved to {csv_path}")

if __name__ == "__main__":
    main()
```
