```python
import itertools
import requests

def generate_urls(base_url, date, first_letter='', second_letter=''):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    first_choices = [first_letter] if first_letter else letters
    second_choices = [second_letter] if second_letter else letters
    
    for combo in itertools.product(first_choices, second_choices):
        url = f"{base_url}{date}{''.join(combo)}.pdf"
        yield url

def test_urls(base_url, date, first_letter='', second_letter=''):
    for url in generate_urls(base_url, date, first_letter, second_letter):
        try:
            response = requests.get(url, allow_redirects=True, timeout=5)
            if response.status_code == 200 and 'application/pdf' in response.headers.get('Content-Type', ''):
                print(f"Valid PDF URL found: {url}")
                return url  # Exit after finding the first successful URL
        except requests.RequestException as e:
            print(f"Error checking {url}: {e}")
            pass
    print("No valid PDF URLs found.")
    return None

if __name__ == "__main__":
    base_url = "https://www.kotaksecurities.com/pdf/indiadaily/indiadaily"
    date = "10082025"  # Change this to the required date in ddmmyyyy format
    first_letter = ""  # Specify first letter if known, else keep blank
    second_letter = ""  # Specify second letter if known, else keep blank
    test_urls(base_url, date, first_letter, second_letter)
```
