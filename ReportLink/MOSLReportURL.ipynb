```python
import requests

def generate_urls(base_url, date, start_page=None, end_page=None):
    start_page = start_page if start_page is not None else 5
    end_page = end_page if end_page is not None else 100
    
    for page in range(start_page, end_page + 1):
        page_str = f"{page:03d}"  # Formats the number to 3-digit format (e.g., 005, 010, 100)
        url = f"{base_url}{date}-MOSL-MI-PG{page_str}.pdf"
        yield url

def test_urls(base_url, date, start_page=None, end_page=None):
    for url in generate_urls(base_url, date, start_page, end_page):
        try:
            response = requests.get(url, allow_redirects=True, timeout=5)
            if response.status_code == 200 and 'application/pdf' in response.headers.get('Content-Type', ''):
                print(f"Valid PDF URL found: {url}")
                return url  # Exit after finding the first successful URL
        except requests.RequestException as e:
            print(f"Error checking {url}: {e}")
            pass
    print("No valid PDF URLs found.")
    return None

if __name__ == "__main__":
    base_url = "https://ftp.motilaloswal.com/emailer/Research/MORNING_INDIA-"
    date = "20251110"  # Change this to the required date in yyyymmdd format
    start_page = 80  # Specify start page if known, else keep None
    end_page = 99  # Specify end page if known, else keep None
    test_urls(base_url, date, start_page, end_page)
```
